```python
#clean dataset to remove columns use drop()method
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
#add column Total that sums up total immigrants by country
df_can['Total'] = df_can.sum(axis = 1)
```

```python
#set column as index: set_index()
df_can.set_index('Country', inplace=True)
```

```python
df_can.head(3)
```

```python
#remove the name of index
df_can.index.name = None
```

```python
#View number of immigrants from Japan(row87)
#1.full row data
print(df_can.loc['Japan'])
```

```python
#alternate methods
print(df_can.iloc[87])
print(df_can[df_can.index == 'Japan'].T.squeeze())
```

```python
#2.for year 2013
print(df_can.loc['Japan', 2013])
```

```python
#alternative method
print(df_can.iloc[87,36])
```

```python
#3.for years 1980 to 1985
print(df_can.loc['Japan',[1980,1981,1982,1983,1984,1985]])
```

```python
#alternative method
print(df_can.iloc[87, [3, 4, 5, 6, 7, 8]])
```

```python
#convert column names into strings;
#[print (type(x)) for x in df_can.columns.values] #<-- uncomment to check type of column headers
df_can.columns = list(map(str, df_can.columns))
```

```python
#Question5
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
#get the top 5 entries
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
print(df_top5)
```
